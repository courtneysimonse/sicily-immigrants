## Formatting Passenger Data

```python
import pandas as pd
import numpy as np
```

```python
# open data file
csv = open('../data/sicily_passengers.csv')
df = pd.read_csv(csv)
```

```python
df.head()
```

```python
df['LastName'] = df['LastName'].str.title()
```

```python
df['FirstName'] = df['FirstName'].str.title()
df['Occupation'] = df['Occupation'].str.title()
df['Literacy'] = df['Literacy'].str.title()
df['origin_city'] = df['origin_city'].str.title()
df['destination_city'] = df['destination_city'].str.title()
```

```python
df.head()
```

```python
df.to_csv('../data/sicily_passengers.csv')
```
