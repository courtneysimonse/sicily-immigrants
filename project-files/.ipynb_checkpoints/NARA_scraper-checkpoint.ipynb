```python
# extract table
table = soup.find(id='queryResults')
#print(table)
```

```python
list_of_rows = []
for row in table.findAll('tr')[2:]:
    list_of_cells = []
    for cell in row.findAll('td')[1:]:
        text = cell.text.replace('age ', '') #remove 'age ' from age field
        list_of_cells.append(text)
    list_of_rows.append(list_of_cells)

print(list_of_rows)
```

```python
# Creation of csv file
outfile = open("data/ellis_island_tes3b3.csv", "w", newline='')
writer = csv.writer(outfile)
writer.writerow(["LastName", "FirstName", "Age", "Occupation", "Literacy", "CountyofOrigin", "CityTownofLastResidence", "DestinationCityTown", "TransitTravelCompartment", "ManifestID"])
writer.writerows(list_of_rows)
```

```python
list_of_rows = []
for i in range(10):
    #next_link_to_click = browser.find_by_xpath(next_link_xpath)
    #next_link_to_click.click()
    browser.click_link_by_partial_text('Next')
    browser.is_text_present('Manifest Identification Number', wait_time=30)
    time.sleep(60)
    # HTML of page
    soup = BeautifulSoup(browser.driver.page_source)

    # extract table
    table = soup.find(id='queryResults')

    for row in table.findAll('tr')[2:]:
        list_of_cells = []
        for cell in row.findAll('td')[1:]:
            text = cell.text.replace('age ', '') #remove 'age ' from age field
            list_of_cells.append(text)
        list_of_rows.append(list_of_cells)
        
    print(browser.url)
    
print(list_of_rows)
```
