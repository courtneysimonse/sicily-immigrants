```python
# extract table
table = soup.find(id='queryResults')
#print(table)
```

```python
# test one page
list_of_rows = []
for row in table.findAll('tr')[2:]:
    list_of_cells = []
    for cell in row.findAll('td')[1:]:
        text = cell.text.replace('age ', '') #remove 'age ' from age field
        list_of_cells.append(text)
    list_of_rows.append(list_of_cells)

print(list_of_rows)
```

```python
# Creation of csv file
outfile = open("data/ellis_island_test3c.csv", "w", newline='')
writer = csv.writer(outfile)
writer.writerow(["LastName", "FirstName", "Age", "Occupation", "Literacy", "CountyofOrigin", "CityTownofLastResidence", "DestinationCityTown", "TransitTravelCompartment", "ManifestID"])
writer.writerows(list_of_rows)
```
