```python
csv3 = open("./data/sicily_passengers.csv")
df3 = pd.read_csv(csv3)
```

```python
df3.axes
```

```python
df3['latlong'].nunique()
dest_list = df3['latlong'].unique()
```

```python
dest_list_df = pd.DataFrame(dest_list)
# rename DestRaw to match passenger df
dest_list_df.rename(columns={0:'latlong'}, inplace=True)
```

```python
dest_list_df.axes
```

```python
# match destination lat/lng to name of destination
dest_list = df3.merge(dest_list_df, how='left', on='latlong')
dest_list.sample(10)
```

```python
pd.DataFrame(dest_list).to_csv('./data/destinations_latlong.csv', index="none")
```
