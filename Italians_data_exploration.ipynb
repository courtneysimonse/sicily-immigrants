```python
import pandas as pd
import numpy as np
```

```python
# open the three passenger data files
csv1 = open('./data/italians_parta.csv', 'r', newline='')
csv2 = open('./data/italians_partb.csv', 'r', newline='')
csv3 = open('./data/italians_partc.csv', 'r', newline='')
```

```python
# create pandas dataframes from the csv files
df1 = pd.read_csv(csv1)
df2 = pd.read_csv(csv2)
df3 = pd.read_csv(csv3)
```

```python
# combine the three passenger files
df = df1.append([df2,df3], sort=False)
df.head()
```

```python
# get an idea of the towns listed as previous last residence
df['CityTownofLastResidence'].value_counts()
```

```python
# get an idea of the destination towns listed
df['DestinationCityTown'].value_counts()
```

```python
# look at Manifest IDs
df['ManifestID'].value_counts()
```

```python
# number of unique Manifest IDs
df['ManifestID'].nunique()
```

```python
# list of Communes with assoicated Province in Sicily
communes = pd.read_csv('./data/communes_sicily.csv')
```

```python
communes.head()
```

```python
# make commune name uppercase to match passenger data
communes['Commune'] = communes['Commune'].str.upper();
```

```python
df['CityTownofLastResidence'].isin(communes['Commune']).value_counts() #39,725 immigrants from Sicily
```

```python
# create new df of only Sicilian passengers by matching reported last residence to Sicily Communes list
sicilydf = df[df['CityTownofLastResidence'].isin(communes['Commune']) == True]
sicilydf
```

```python
# match Commune with previous residence to add Province column
sicilydf = pd.merge(sicilydf,communes,how='left',left_on='CityTownofLastResidence',right_on='Commune')
```

```python
# drop population and duplicate Commune columns
sicilydf = sicilydf.drop(['Population','Commune'],axis='columns')
```

```python
df['CityTownofLastResidence'].isna().value_counts()
```

```python
sicilydf.head()
```

```python
sicilydf['Province'].value_counts()
```

```python
df['Literacy'].value_counts()
```

```python
# write all passenger data to csv
df.to_csv('./data/italians_to_america.csv')
```

```python
# create dataframe of manifest header data csv
manifests = pd.read_csv('data/manifest_list_1855_1900.csv')
manifests.head()
```

```python
# convert Manifest ID in sicilydf to int to match manifest header file
sicilydf['ManifestID'] = sicilydf['ManifestID'].astype('int')
```

```python
# add manifest header information to passenger dataframe
sicilydf = pd.merge(sicilydf, manifests, how='left', on='ManifestID')
```

```python
sicilydf.head()
```

```python
sicilydf['Age'].value_counts()
```

```python
# fix typo in Country of Origin column
sicilydf.rename(columns={'CountyofOrigin':'CountryofOrigin'}, inplace=True)
```

```python
# bring in destinations lat/lng from geocoding notebook
csv = open('./data/geocode_results.csv')
dest_df = pd.read_csv(csv)
dest_df.head()
```

```python
# rename DestRaw to match passenger df
dest_df.rename(columns={'DestRaw':'DestinationCityTown'}, inplace=True)
```
